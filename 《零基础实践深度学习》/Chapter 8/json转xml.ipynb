```python
import json
import os
from xml.etree.ElementTree import Element
from xml.etree.ElementTree import tostring
import xml.dom.minidom

#生成xml文件
def GenerateXml(dict,xml_path):
    impl = xml.dom.minidom.getDOMImplementation()
    dom = impl.createDocument(None, 'annotation', None)
    
    root = dom.documentElement

    FOLDER = dom.createElement('folder')
    FOLDER.appendChild(dom.createTextNode(dict['folder']))
    root.appendChild(FOLDER)

    FILENAME = dom.createElement('filename')
    FILENAME.appendChild(dom.createTextNode(dict['filename']))
    root.appendChild(FILENAME)

    FILEPATH = dom.createElement('path')
    FILEPATH.appendChild(dom.createTextNode(dict['path']))
    root.appendChild(FILEPATH)

    SOURCE = dom.createElement('source')
    DATABASE = dom.createElement('database')
    DATABASE.appendChild(dom.createTextNode('Unknown'))
    SOURCE.appendChild(DATABASE)
    root.appendChild(SOURCE)

    SIZE = dom.createElement('size')
    WIDTH = dom.createElement('width')
    HEIGHT = dom.createElement('height')
    DEPTH = dom.createElement('depth')
    WIDTH.appendChild(dom.createTextNode('640'))
    HEIGHT.appendChild(dom.createTextNode('240'))
    DEPTH.appendChild(dom.createTextNode('3'))
    SIZE.appendChild(WIDTH)
    SIZE.appendChild(HEIGHT)
    SIZE.appendChild(DEPTH)
    root.appendChild(SIZE)

    for i in range(len(dict['objects'])):
        obj = dict['objects'][i]

        OBJECT = dom.createElement('object')
        NAME = dom.createElement('name')
        POSE = dom.createElement('pose')
        TRUNCATED = dom.createElement('truncated')
        DIFFICULT = dom.createElement('difficult')

        NAME.appendChild(dom.createTextNode(obj['name']))
        POSE.appendChild(dom.createTextNode(obj['pose']))
        TRUNCATED.appendChild(dom.createTextNode('0'))
        DIFFICULT.appendChild(dom.createTextNode('0'))

        OBJECT.appendChild(NAME)
        OBJECT.appendChild(POSE)
        OBJECT.appendChild(TRUNCATED)
        OBJECT.appendChild(DIFFICULT)

        BNDBOX = dom.createElement('bndbox')

        XMIN = dom.createElement('xmin')
        YMIN = dom.createElement('ymin')
        XMAX = dom.createElement('xmax')
        YMAX = dom.createElement('ymax')

        XMIN.appendChild(dom.createTextNode(obj['xmin']))
        YMIN.appendChild(dom.createTextNode(obj['ymin']))
        XMAX.appendChild(dom.createTextNode(obj['xmax']))
        YMAX.appendChild(dom.createTextNode(obj['ymax']))
        
        BNDBOX.appendChild(XMIN)
        BNDBOX.appendChild(YMIN)
        BNDBOX.appendChild(XMAX)
        BNDBOX.appendChild(YMAX)

        OBJECT.appendChild(BNDBOX)

        root.appendChild(OBJECT)


    f= open(xml_path, 'w') #w替换为a，追加
    dom.writexml(f, addindent=' ', newl='\n')
    f.close()

#函数输入是(json文件路径，存储xml的文件夹)
def json_to_xml(json_path,xml_folder_path):
    with open(json_path,'r',encoding='utf8')as fp:
        json_data = json.load(fp)
        json_data['imageData'] = None
        folder = 'wuke'
        filename = str(json_path.strip().split('/')[-1]).split('.')[0] + '.jpeg'
        path = './annotations/xmls/' + folder
        width = str(json_data['imageWidth'])
        height = str(json_data['imageHeight'])
        depth = '3'
        num_object = len(json_data['shapes'])
        objects = []
        for i in range(num_object):
            points = json_data['shapes'][i]['points']
            objects.append({'name':json_data['shapes'][i]['label'],'pose':'Unspecified','truncated':0,'difficult':0,'xmin':str(int(min(points[0][0],points[1][0]))),'ymin':str(int(min(points[0][1],points[1][1]))),'xmax':str(int(max(points[0][0],points[1][0]))),'ymax':str(int(max(points[0][1],points[1][1])))})
        dict = {'folder':folder,'filename':filename,'path':path,'width':width,'height':height,'depth':depth,'objects':objects}
        xml_path = xml_folder_path + '/' + json_path.split('/')[-1].split('.')[0] + '.xml'
        GenerateXml(dict,xml_path)
        print(xml_path)
```

```python


filepath = 'C:/Users/吴坷/Desktop/insect_det/JPEGImages' 
filenames = os.listdir(filepath)
for filename in filenames:
    if filename.split('.')[-1] == 'json':
        json_to_xml(filepath+'/'+filename,'C:/Users/吴坷/Desktop/xmls')
```
